```python
import pandas as pd
import numpy as np
import datetime as dt
from statsmodels.tsa.seasonal import STL
import matplotlib.pyplot as plt

# -------------------------------
# 1. Generate Synthetic Toddler Data
# -------------------------------
def generate_toddler_data_spring_forward(start_date="2024-01-01", end_date="2024-03-31"):
    dates = pd.date_range(start=start_date, end=end_date, freq="D")
    np.random.seed(42)
    
    # DST shift: Spring Forward
    dst_shift = dt.datetime(2024, 3, 10)
    
    # Daylight: deterministic linear increase from 9h (Jan) to 12h (Mar)
    dl_start = 9 * 60    # minutes
    dl_end   = 12 * 60   # minutes
    daylight_min = np.linspace(dl_start, dl_end, len(dates))
    
    # Apply DST shift: reduce morning light effect
    dst_index = np.where(dates == dst_shift)[0][0]
    daylight_min[dst_index:] += 60  # morning feels "later"
    
    # Initialize lists
    outdoor_play = []
    indoor_play = []
    meltdowns = []
    interaction_reqs = []
    nap_duration = []
    wake_times = []
    bed_times = []
    field_trip = []
    field_location = []
    
    for i, d in enumerate(dates):
        dl_min = daylight_min[i]
        is_after_dst = d >= dst_shift
        
        # Outdoor play proportional to daylight
        base_outdoor = np.random.normal(dl_min * 1.75, 20)
        if is_after_dst:
            base_outdoor *= np.random.uniform(0.7, 1.0)  # less outdoor play shock
        base_outdoor = max(base_outdoor, 0)
        outdoor_play.append(base_outdoor)
        
        # Indoor play compensates
        indoor_base = np.random.normal(20, 10) + max(0, (600 - dl_min))
        if is_after_dst:
            indoor_base *= np.random.uniform(1.0, 1.2)
        indoor_play.append(indoor_base)
        
        # Nap duration declines post-DST
        nap_base = np.random.normal(130, 10)
        if is_after_dst:
            nap_base -= np.random.uniform(15, 30)  # harder to nap
        nap_duration.append(max(nap_base, 0))
        
        # Meltdowns increase slightly after DST
        meltdown_rate = 1 + max(0, (90 - base_outdoor) / 30)
        if is_after_dst:
            meltdown_rate *= np.random.uniform(1.1, 1.3)
        meltdowns.append(max(0, int(np.random.poisson(meltdown_rate))))
        
        # Interaction requests rise with indoor play
        interaction_rate = 5 + indoor_base / 25
        if is_after_dst:
            interaction_rate *= 1.15
        interaction_reqs.append(int(np.random.poisson(interaction_rate)))
        
        # Sleep schedule
        wake_shift = 20 if is_after_dst else 0   # wakes later
        bed_shift  = -15 if is_after_dst else 0  # goes to bed slightly earlier
        wake_times.append(
            (dt.datetime(2024,1,1,7,0) + dt.timedelta(minutes=np.random.normal(wake_shift, 20))).time()
        )
        bed_times.append(
            (dt.datetime(2024,1,1,20,0) + dt.timedelta(minutes=np.random.normal(bed_shift, 30))).time()
        )
        
        # Field trips: ~3/7 chance
        trip = np.random.rand() < 3/7
        field_trip.append(trip)
        if trip:
            field_location.append(np.random.choice(["Museum", "Grocery", "Library", "Park", "Aquarium"]))
        else:
            field_location.append(None)
    
    # Build DataFrame
    df = pd.DataFrame({
        "date": dates,
        "daylight_min": daylight_min,
        "outdoor_play_min": outdoor_play,
        "indoor_play_min": indoor_play,
        "meltdown_count": meltdowns,
        "interaction_requests": interaction_reqs,
        "nap_duration_min": nap_duration,
        "wake_time": wake_times,
        "bed_time": bed_times,
        "field_trip": field_trip,
        "field_location": field_location
    })
    
    return df

# Generate dataset
df = generate_toddler_data_spring_forward()

# -------------------------------
# 2. Normalize Variables
# -------------------------------
normalized = df.copy()
cols_to_norm = [
    "outdoor_play_min",
    "indoor_play_min",
    "meltdown_count",
    "interaction_requests",
    "nap_duration_min",
    "daylight_min"
]

for col in cols_to_norm:
    min_val = df[col].min()
    max_val = df[col].max()
    normalized[col + "_norm"] = (df[col] - min_val) / (max_val - min_val)

# -------------------------------
# 3. STL Decomposition & Trend Plots
# -------------------------------
res_out = STL(normalized['outdoor_play_min_norm'], period=7).fit()
res_in  = STL(normalized['indoor_play_min_norm'], period=7).fit()
res_nap = STL(normalized['nap_duration_min_norm'], period=7).fit()
res_int = STL(normalized['interaction_requests_norm'], period=7).fit()
res_melt = STL(normalized['meltdown_count_norm'], period=7).fit()
res_day = normalized['daylight_min_norm']  # deterministic

plt.figure(figsize=(12,8))
plt.plot(normalized['date'], res_nap.trend, label="Nap Duration", linewidth=3)
plt.plot(normalized['date'], res_out.trend, label="Outdoor Play")
plt.plot(normalized['date'], res_in.trend, label="Indoor Play")
plt.plot(normalized['date'], res_int.trend, label="Interaction Requests")
plt.plot(normalized['date'], res_day, label="Available Daylight")
plt.plot(normalized['date'], res_melt.trend, label="Meltdown Count")
plt.axvline(pd.to_datetime("2024-03-10"), color='black', linestyle='--', alpha=0.5, label="DST Start")
plt.xticks(pd.date_range(df['date'].min(), df['date'].max(), freq='MS'),
           [d.strftime('%b %Y') for d in pd.date_range(df['date'].min(), df['date'].max(), freq='MS')],
           rotation=45)
plt.title("Normalized Toddler Activity Trends: Spring Forward DST")
plt.xlabel("Date")
plt.ylabel("Normalized Value")
plt.legend()
plt.tight_layout()
plt.show()

# -------------------------------
# 4. Nap Decomposition Plot
# -------------------------------
fig = res_nap.plot()
axes = fig.axes
axes[0].set_title("Observed Nap Duration")
axes[1].set_title("Trend in Nap Duration")
axes[2].set_title("Nap Duration Seasonality")
axes[3].set_title("Residuals")
plt.tight_layout()
plt.show()
```
